```python
label_bbox = {}

for value in unique_values:
    contents = value.split(",")
    
    # remove zero after decimal and only keep integer
    for i in range(1, len(contents)):
        if int(float(contents[i])) - float(contents[i]) == 0.0:
            contents[i] = str(int(float(contents[i])))
    
    label_bbox[contents[0]] = contents[1:]
    
    
label_bbox
print(type(label_bbox["Basic Information___First Name"][0]))
```

```python
if sorted(label_bbox.keys()) == sorted(labels_in_order):
    print("Arrays have the same elements.")
else:
    print("Arrays are different.")
```

```python
import fitz  # PyMuPDF

def extract_text_from_bbox(pdf_path, page_number, bbox):
    pdf_document = fitz.open(pdf_path)
    page = pdf_document[page_number]
    
    # the given bbox are highly scaled
    # PyMuPDF displays page size as Rect(0.0, 0.0, 612.0, 792.0)
    # but bbox in train dataset is maximum of 1600.8 wide
    # so scaling_factor used to sacle the bbox appropriately for PyMuPDF
    rect = page.rect
    # rect[2] is maximum width of opened page
    scaling_factor = rect[2] / 1700.0

    left, top, right, bottom = [x * scaling_factor for x in bbox]

    # Create a rectangle from the bbox
    rect = fitz.Rect(left, top, right, bottom)

    # Get the text in the specified bbox
    text = page.get_text("text", clip=rect)
    pdf_document.close()

    return text
```

```python
def extract_label_value(pdf_path, label):
    page_number = 0
    bbox_coordinates = [float(x) for x in label_bbox[label]]
    
    extracted_text = extract_text_from_bbox(pdf_path, page_number, bbox_coordinates)
    return extracted_text.strip()
```
